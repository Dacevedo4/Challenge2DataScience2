```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

Conclusiones del análisis de Churn

La tasa general de evasión es del 25,6%, lo que representa un porcentaje crítico que requiere atención estratégica.

La mayoría de los abandonos ocurre en los primeros meses de permanencia, indicando que los clientes nuevos son los más vulnerables.

Los contratos mensuales presentan tasas de abandono significativamente más altas. Aunque promover únicamente contratos a largo plazo podría ser difícil por la competencia y regulaciones locales, es recomendable mejorar la experiencia y beneficios de los contratos mensuales, de modo que los clientes se sientan satisfechos y retenerlos dentro del plan.

Los clientes con contratos a largo plazo muestran menor probabilidad de abandono, evidenciando que la fidelización funciona cuando los clientes se comprometen por más tiempo.

La fibra óptica representa un segmento de riesgo elevado cuando no se combina con servicios de soporte técnico, por lo que ofrecer paquetes completos puede reducir la evasión.

Los clientes que no utilizan servicios complementarios como seguridad online, backup en la nube o protección de dispositivos tienen un mayor registro de cancelaciones.

Los métodos de pago manuales o no automáticos se asocian a una mayor probabilidad de evasión, lo que sugiere que incentivar pagos automáticos puede mejorar la retención.

Los clientes con tarifas más altas tienden a abandonar más, posiblemente debido a la percepción de costo-beneficio.

En general, los servicios adicionales actúan como barreras frente a la cancelación y contribuyen a mejorar la retención de clientes.
